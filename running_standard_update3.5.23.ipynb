```python
import mysql.connector
from chatgpt_wrapper import ChatGPT
import nest_asyncio
nest_asyncio.apply()
import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import askyesno, askquestion
from tkinter import *
import smtplib
from email.message import EmailMessage
from email.mime.text import MIMEText
```

```python
def __main__(): 
    window=tk.Tk()
    window.geometry("800x800")
    window.title("City Group Inputs")
    
    location_name = tk.Label(text = "Enter location")
    location_name.grid(column=0,row=1)
    
    manager_name = tk.Label(text = "Enter your name")
    manager_name.grid(column=0,row=2)
    
    customer_name = tk.Label(text = "Enter the name of the customer")
    customer_name.grid(column=0,row=3)
    
    order_name = tk.Label(text = "Enter order")
    order_name.grid(column=0,row=4)
    
    descriptor_1 = tk.Label(text = "Please enter the first descriptor")
    descriptor_1.grid(column=0,row=5)
    
    descriptor_2 = tk.Label(text = "Please enter the second descriptor")
    descriptor_2.grid(column=0,row=6)
    
    descriptor_3 = tk.Label(text = "Please enter the third descriptor")
    descriptor_3.grid(column=0,row=7)
    
    mail = tk.Label(text = "Enter the customer's email")
    mail.grid(column=0,row=8)
    
    message = tk.Label(text = "Upon entering, you will be prompted to approve or deny the message to be sent")
    message.grid(column=0,row=9)
    
    loc_Entry = tk.Entry()
    loc_Entry.grid(column=1,row=1)
    
    man_Entry = tk.Entry()
    man_Entry.grid(column=1,row=2)
    
    cust_Entry = tk.Entry()
    cust_Entry.grid(column=1,row=3)
    
    order_Entry = tk.Entry()
    order_Entry.grid(column=1,row=4)
    
    desc1_Entry = tk.Entry()
    desc1_Entry.grid(column=1,row=5)
    
    desc2_Entry = tk.Entry()
    desc2_Entry.grid(column=1,row=6)
    
    desc3_Entry = tk.Entry()
    desc3_Entry.grid(column=1,row=7)
    
    mail_Entry = tk.Entry()
    mail_Entry.grid(column=1,row=8)
    
    #Establishes sending email and credentials, sets message and sends to specified addresss.
    def mail(resp,location,mail):
        email_address = "citytester4125@gmail.com"
        email_password = ""
        msg = EmailMessage()
        msg.set_content("Thank you for dining at {}! Please consider leaving a review, we have generated one based on the feedback given to our manager by you.\n{}".format(location,MIMEText(resp)))
        msg['Subject'] = "City Group Hospitality would like to thank you!"
        msg['From'] = "citytester4125@gmail.com"
        msg['To'] = mail
    
        with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp:
            smtp.login(email_address, email_password)
            smtp.send_message(msg)
    
    #MySQL Connection and insertion of response,order,customer name,and location.
    def approve(response,loc,order,name):
        cnx = mysql.connector.connect(user='root',password='Tootie2001', database='target')
        cursor = cnx.cursor()
        query =("INSERT INTO mealExp(feedback,itemOrdered,custName,location) VALUES (%(x)s,%(y)s,%(z)s,%(w)s)")
        values = {'x':response,'y':order,'z':name,'w':loc}
        cursor.execute(query,values)
        cursor.close()
        cnx.commit()
        cnx.close()
    
     #Confirms sending address following approval.
    def confirmation(mail):
        messagebox.showinfo("Mail Sent", "Sent to {}".format(mail))
    
    #Confirm user would like to write to database.
    def database(response,loc,order,cust):
        dataprove = tk.messagebox.askyesno(title='Write to database?', message='Add to MySQL Database?')
        if dataprove:
            approve(response,loc,order,cust)
            print(cust)
        else:
            print('did not write')
    
    #Input of values and calling of auxilliary prodcution functions. Establishes connection to ChatGPT()instance.
    def operate():
        desc_list = [loc_Entry.get(),desc1_Entry.get(),desc2_Entry.get(),desc3_Entry.get()]
        bot = ChatGPT()
        response = bot.ask("Write a short one paragraph review for a restaurant named {} based on these keywords {},{},{}".format(*(x for x in desc_list)))
        approval = tk.messagebox.askyesno(title='Approval', message=response)
        if approval:
            database(response,loc_Entry.get(),order_Entry.get(),cust_Entry.get())
            mail(response,str(loc_Entry.get()),str(mail_Entry.get()))
            confirmation(mail_Entry.get())
        else:
            pass

    #Remove text from Entries.
    def clear_text():
        loc_Entry.delete(0, END)
        cust_Entry.delete(0, END)
        order_Entry.delete(0,END)
        desc1_Entry.delete(0, END)
        desc2_Entry.delete(0, END)
        desc3_Entry.delete(0, END)
        mail_Entry.delete(0, END)
    
    button=tk.Button(window,text="Run",command = lambda:[operate(),clear_text()])
    button.grid(column=1,row=9)
    destroy=tk.Button(window, text ="Quit", command = window.destroy)
    destroy.grid(column=1,row=10)
    
    window.mainloop()

if __name__ == '__main__':
    __main__()
```
